```python
# This shows the month & year wise together & sort values by month.

df1['month'] = pd.to_datetime(df1['datetime']).dt.to_period('M')
monthlyTransactionForItem = df1[['month','Transaction', 'Item']].groupby(['month', 'Item'], as_index=False).count().sort_values(by='month')
monthlyTransactionForItem.set_index('month' ,inplace=True)
```

```python
# Market basket Analysis

from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
```

```python
transaction_list = []

# For loop to create a list of the unique transactions throughout the dataset.
for i in df1['Transaction'].unique():
    tlist = list(set(df1[df1['Transaction']==i]['Item']))
    if len(tlist)>0:
        transaction_list.append(tlist)
print(len(transaction_list))
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transaction_list).transform(transaction_list)
df2 = pd.DataFrame(te_ary, columns = te.columns_)
```
